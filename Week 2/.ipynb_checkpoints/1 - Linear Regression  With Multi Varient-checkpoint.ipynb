## Model Building

```python
def compute_cost(X, y, W, b):
    m = len(X)
    f_wb = np.dot(X, W) + b
    cost = np.sum((f_wb - y)**2)
    cost = (1/(2*m))*cost
    
    return cost
```

```python
def gradient_descent(X, y, W, b, alpha):
    m = len(X)
    iteration = 10000
    
    for i in range(iteration):
        f_wb = np.dot(X, W) + b
        
        cost = compute_cost(X, y, W, b)
        print(f"Iteration = {i}, Cost = {cost}")
        
        d_dw = (1/m)*np.sum(np.dot((f_wb - y),X))
        d_db = (1/m)*np.sum(f_wb - y)
        
        W = W - alpha * d_dw
        b = b - alpha * d_db
    
    return W, b
```

```python
y = train_df["Temperature (C)"][:2000]
X = train_df.drop(["Temperature (C)"], axis=1)[:2000]
W = np.zeros(X.shape[1])
b = 0
alpha = 0.01

W_updated, b_updated = gradient_descent(X, y, W, b, alpha)
```

## Predicting

```python
def predict(data_point, W, b):
    return np.dot(data_point, W) + b
```

```python
predict(test_df.iloc[0], W_updated, b_updated)
```
